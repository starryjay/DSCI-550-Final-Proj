```python
import pandas as pd

diabetes = pd.read_csv("DSCI-550-Final-Proj/diabetes_binary_5050split_health_indicators_BRFSS2015.csv")
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
diabetes
```

```python
from keras.models import Sequential
from keras.layers import Dense

X = diabetes.loc[:, diabetes.columns!='Diabetes_binary']
y = diabetes['Diabetes_binary']
model = Sequential()
model.add(Dense(21, input_shape=(21,), activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=150, batch_size=10)
```

```python
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```
