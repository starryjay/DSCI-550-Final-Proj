# Fitting an Artificial Neural Network using Keras

## Importing Data

```python
import pandas as pd
diabetes = pd.read_csv("diabetes_binary_5050split_health_indicators_BRFSS2015.csv")
```

```python
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
diabetes
```

## First Model

Let's try fitting a basic Sequential neural network with Dense layers on this data. We'll graph accuracy, ROC-AUC, precision, and recall to get a good idea of the model's fit.

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout

X = diabetes.loc[:, diabetes.columns!='Diabetes_binary']
y = diabetes['Diabetes_binary']
model = Sequential()
model.add(Dense(21, input_shape=(21,), activation='relu'))
model.add(Dense(1, activation='sigmoid'))
adam = tf.keras.optimizers.legacy.Adam(learning_rate=0.01)
model.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy', 'AUC', 'Precision', 'Recall'])
```

```python
history = model.fit(X, y, validation_split=0.25, epochs=150, batch_size=100)
```

```python
loss, accuracy, auc, precision, recall = model.evaluate(X, y)
print('Loss: %.2f' % (loss*100))
print('Accuracy: %.2f' % (accuracy*100))
print('AUC: %.2f' % (auc*100))
print('Precision: %.2f' % (precision*100))
print('Recall: %.2f' % (recall*100))
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

plt.plot(history.history['auc'])
plt.plot(history.history['val_auc'])
plt.title('model AUC')
plt.ylabel('auc')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

plt.plot(history.history['precision'])
plt.plot(history.history['val_precision'])
plt.title('model precision')
plt.ylabel('precision')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

plt.plot(history.history['recall'])
plt.plot(history.history['val_recall'])
plt.title('model recall')
plt.ylabel('recall')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

There seem to be some errors with the fit of the model, given these strange metric graphs. 



## Second Model

Let's try fitting a model with more layers and a different number of nodes.

```python
model2 = Sequential()
model2.add(Dense(64, input_shape=(21,), activation='relu'))
model2.add(Dense(32, activation='relu'))
model2.add(Dense(1, activation='sigmoid'))
model2.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy', 'AUC', 'Precision', 'Recall'])
```

```python
history_2 = model2.fit(X, y, validation_split=0.25, epochs=150, batch_size=100)
```

```python
plt.plot(history_2.history['accuracy'])
plt.plot(history_2.history['val_accuracy'])
plt.title('second model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

plt.plot(history_2.history['auc'])
plt.plot(history_2.history['val_auc'])
plt.title('second model AUC')
plt.ylabel('auc')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

plt.plot(history_2.history['precision'])
plt.plot(history_2.history['val_precision'])
plt.title('second model precision')
plt.ylabel('precision')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

plt.plot(history_2.history['recall'])
plt.plot(history_2.history['val_recall'])
plt.title('second model recall')
plt.ylabel('recall')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

This doesn't seem to have changed anything. 

## Third Model

Let's try normalizing the predictors before we compile the model this time.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X = scaler.fit_transform(X)

model3 = Sequential()
model3.add(Dense(21, input_shape=(21,), activation='relu'))
model3.add(Dense(1, activation='sigmoid'))
model3.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy', 'AUC', 'Precision', 'Recall'])
history_3 = model3.fit(X, y, validation_split=0.25, epochs=150, batch_size=100)
```

```python
plt.plot(history_3.history['accuracy'])
plt.plot(history_3.history['val_accuracy'])
plt.title('third model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

plt.plot(history_3.history['auc'])
plt.plot(history_3.history['val_auc'])
plt.title('third model AUC')
plt.ylabel('auc')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

plt.plot(history_3.history['precision'])
plt.plot(history_3.history['val_precision'])
plt.title('third model precision')
plt.ylabel('precision')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

plt.plot(history_3.history['recall'])
plt.plot(history_3.history['val_recall'])
plt.title('third model recall')
plt.ylabel('recall')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

Now the test accuracy is improving a bit more, though the other metric graphs still look incorrectly fit. 


## Fourth Model

Let's try splitting `X` and `y` into a training and testing set beforehand instead of using the `validation_split` parameter to see if this helps. We'll still normalize inputs since that did lead to some improvement last time. We'll also use the same model.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, shuffle=True)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
history_4 = model3.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=150, batch_size=100)
```

```python
loss, accuracy, auc, precision, recall = model3.evaluate(X_test, y_test)

print("Loss:", loss*100)
print("Accuracy:", accuracy*100)
print("AUC:", auc*100)
print("Precision:", precision*100)
print("Recall:", recall*100)
```

```python
plt.plot(history_4.history['accuracy'])
plt.plot(history_4.history['val_accuracy'])
plt.title('fourth model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

plt.plot(history_4.history['auc'])
plt.plot(history_4.history['val_auc'])
plt.title('fourth model AUC')
plt.ylabel('auc')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

plt.plot(history_4.history['precision'])
plt.plot(history_4.history['val_precision'])
plt.title('fourth model precision')
plt.ylabel('precision')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

plt.plot(history_4.history['recall'])
plt.plot(history_4.history['val_recall'])
plt.title('fourth model recall')
plt.ylabel('recall')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

This looks *much* better than our previous examples. While ROC-AUC is decreasing over time for the testing set, which is not ideal, it still hovers between 0.826 and 0.833, meaning that the model has good predictive power. The ROC-AUC score is also increasing over time for the training set, which is a good sign.

## Fifth Model

Let's try to improve the accuracy and ROC-AUC a bit more. We'll add a Dropout layer to our model to account for any overfitting, setting the dropout rate to 50%. Let's also reshuffle the training and testing data.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, shuffle=True)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from keras.callbacks import ReduceLROnPlateau, EarlyStopping

learning_reducer = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.0001)
early_stopping = EarlyStopping(monitor='val_loss', min_delta=0.0001, patience=5, restore_best_weights=True)

model5 = Sequential()
model5.add(Dense(21, input_shape=(21,), activation='relu'))
model5.add(Dense(1, activation='sigmoid'))
model5.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy', 'AUC', 'Precision', 'Recall'])
history_5 = model5.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=150, batch_size=100, callbacks=[learning_reducer, early_stopping])
```

```python
plt.plot(history_5.history['accuracy'])
plt.plot(history_5.history['val_accuracy'])
plt.title('fifth model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

plt.plot(history_5.history['auc'])
plt.plot(history_5.history['val_auc'])
plt.title('fifth model AUC')
plt.ylabel('auc')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

plt.plot(history_5.history['precision'])
plt.plot(history_5.history['val_precision'])
plt.title('fifth model precision')
plt.ylabel('precision')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

plt.plot(history_5.history['recall'])
plt.plot(history_5.history['val_recall'])
plt.title('fifth model recall')
plt.ylabel('recall')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
