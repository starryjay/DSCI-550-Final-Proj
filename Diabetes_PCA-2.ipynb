```python
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
x = diabetes_df.loc[:,'HighBP':].values
x = StandardScaler().fit_transform(x) 
x.shape
```

```python
np.mean(x),np.std(x)
```

```python
columns_indic = ['factor'+str(i) for i in range(x.shape[1])]
normalDiabetes_df = pd.DataFrame(x,columns = columns_indic)
```

```python
normalDiabetes_df.head()
```

```python
from sklearn.decomposition import PCA
pca_diabetes = PCA(n_components=14)
p_Components_Diabetes = pca_diabetes.fit_transform(x)
```

```python
pca_diabetes_df = pd.DataFrame(data = p_Components_Diabetes, columns = ['Principal Component 1', 'Principal Component 2','Principal Component 3','Principal Component 4','Principal Component 5','Principal Component 6','Principal Component 7','Principal Component 8','Principal Component 9','Principal Component 10','Principal Component 11','Principal Component 12','Principal Component 13','Principal Component 14'])



```

```python
print('Explained variation per principal component: {}'.format(pca_diabetes.explained_variance_ratio_))
```

```python
import matplotlib.pyplot as plt
```

```python
diabetes_df['Diabetes_binary'].replace(0, 'No Diabetes',inplace=True)
diabetes_df['Diabetes_binary'].replace(1, 'Diabetes/Pre-Diabetes',inplace=True)
```

```python
plt.figure(figsize=(6,6))
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlim(-5, 8) 
plt.ylim(-4, 9) 
plt.xlabel('Principal Component 1',fontsize=18)
plt.ylabel('Principal Component 2',fontsize=18)
plt.title("Principal Component Analysis of Diabetes Dataset",fontsize=20)
targets = ['No Diabetes','Diabetes/Pre-Diabetes']
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = diabetes_df['Diabetes_binary'] == target
    plt.scatter(pca_diabetes_df.loc[indicesToKeep, 'Principal Component 1']
               , pca_diabetes_df.loc[indicesToKeep, 'Principal Component 2'], c = color, s = 50)

plt.legend(targets,prop={'size': 10})
```
